```python
dfValue = timesData
dfValue.income = pd.to_numeric(dfValue.income, errors='coerce')

ax = dfValue.groupby(['country']).income.mean().sort_values(ascending=False).plot(kind='bar', figsize = (18,5))
ax.set_xlabel('Pays', fontsize=14)
ax.set_title("Le score universitaire pour le transfert de connaissances par pays",fontsize=16)
ax.set_ylabel("Score moyen du revenu", fontsize=14)
```

```python
import plotly.graph_objs as go

dfValue = timesData
ax = dfValue.groupby('country').agg(np.mean).index
y1 = dfValue.groupby('country').income.agg(np.mean).sort_values(ascending=False)
y2 =  dfValue.groupby('country').international.mean().sort_values(ascending=False)

trace1 = {
  'x': ax,
  'y': y1, 
  'name': 'income',
  'type': 'bar'
};
trace2 = {
  'x': ax,
  'y': y2,
  'name': 'international',
  'type': 'bar'
};
data = [trace1, trace2];
layout = {
  'xaxis': {'title': 'Top des 3 meilleurs universités'},
  'barmode': 'relative',
  'title': 'Correlation entre le score "income" et le score "international"'
};
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
